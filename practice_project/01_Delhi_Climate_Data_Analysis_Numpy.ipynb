# Delhi_Climate_Data_Analysis_NumPy.ipynb

```python
import numpy as np
```

## Load the data from csv.

```python
data = np.genfromtxt("/content/DailyDelhiClimateTest.csv",delimiter=",",skip_header=1,usecols=1)
```

```python
dates = np.genfromtxt("/content/DailyDelhiClimateTest.csv",delimiter=",",skip_header=1,usecols=0, dtype='str')
```

## Extracting features (columns) into NumPy arrays

```python
temp = data
humidity = np.genfromtxt("/content/DailyDelhiClimateTest.csv", delimiter=",", skip_header=1, usecols=2)
wind = np.genfromtxt("/content/DailyDelhiClimateTest.csv", delimiter=",", skip_header=1, usecols=3)
pressure = np.genfromtxt("/content/DailyDelhiClimateTest.csv", delimiter=",", skip_header=1, usecols=4)
```

## Total Records

```python
print("Total Records:", len(temp))
```

## Temperature Analysis

```python
# Average Temperature
print("Average Temperature:",np.mean(temp))
```

```python
# Maximum Temperature
print("Maximum Temperature:",np.max(temp))
```

```python
# Minimum Temperature
print("Minimum Temperature:",np.min(temp))
```

## Weather Condition Analysis

```python
# Hot Days (>30°C)
print(np.sum(temp > 30))
```

```python
# Days Above Average Temperature
print("Days Above Average Temperature  :", np.mean(temp))
```

## Wind Speed Analysis

```python
# Average Wind Speed
print("Average Wind Speed:",np.mean(wind))
```

```python
# Maximum Wind Speed
print("Maximum Wind Speed:",np.max(wind))
```

## Pressure Analysis

```python
# Maximum Pressure
print("Maximum Pressure:",np.max(pressure))
```

```python
# Average Pressure
print("Average Pressure:",np.mean(pressure))
```

```python
# Minimum Pressure
print("Minimum Pressure:",np.min(pressure))
```

## Correlation Analysis

```python
# Temperature vs Humidity
np.corrcoef(temp,humidity)
```

```python
# Temperature vs Wind Speed
np.corrcoef(temp,wind)
```

```python
# Temperature vs Pressure
np.corrcoef(temp,pressure)
```

## Sorting Analysis

```python
# Top 10 Hottest Temperatures
print(np.sort(temp)[-10:])
```

```python
# Top 10 Coldest Temperatures
print(np.sort(temp)[:10])
```
